```python
import math
def formulas(a, b, c):
    """Formulas to calculate values in standard form
    returns x and y values of vertex, y intercept, x intercepts
    """
    xmin = -b/(2*a)
    ymin = c - (b**2/(4*a))
    discriminant = math.sqrt(b**2-4*a*c)
    xint1 = (-b + discriminant)/(2*a) 
    xint2 = (-b - discriminant)/(2*a)
    yint = c
    return xmin, ymin, xint1, xint2, yint

a,b,c,d,e = formulas(2, -8, 7)
print(a)
xmin = 100
print(xmin)
```

```python
def findymin(a,b,c):
    _, y, _, _, _ = formulas(a, b ,c)
    return y

findymin(1,-6,-1)
```
